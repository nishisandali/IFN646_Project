```python
import sys
import os
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O

# visualisation imports
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# seaborn styling default
sns.set_theme(style='darkgrid')

# hide warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('raw_data/p02_n01_ready.csv')
```

```python
df2 = pd.read_csv('raw_data/p02_n02_ready.csv')
```

```python
frames = [df1, df2]

result = pd.concat(frames)
```

```python
result['Dates'] = pd.to_datetime(result['timestamp']).dt.date
result['Time'] = pd.to_datetime(result['timestamp']).dt.time
```

```python
result = result.drop(['timestamp'], axis=1)
result = result.drop(['Dates'], axis=1)
```

```python
#Since nan values are very few compared to the number of records, 
#all the nan records will be dropped
result = result.dropna()
```
